#Pivot Tables & Advanced Aggregations

Goal: Summarize raw data into a "Report" format instantly.

#The Raw Data

```python
import pandas as pd

# Imagine an E-commerce dataset
data = {
    'Store': ['NY', 'NY', 'LA', 'LA', 'CH', 'CH', 'NY', 'LA'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Electronics'],
    'Revenue': [500, 300, 600, 400, 550, 350, 700, 650]
}
df = pd.DataFrame(data)
```

#The Pivot Table (The "Report" Builder)

Concept: I want to see Category as rows and Store as columns.

```python
# This summarizes exactly where the money is coming from
report = df.pivot_table(index='Category', columns='Store', values='Revenue', aggfunc='sum')

print("Revenue Report by Store & Category:\n", report)
```

#Multi-Agg (The "Stats" Builder)

Concept: I want the Mean, Max, and Total Revenue for each Store.

```python
store_stats = df.groupby('Store')['Revenue'].agg(['mean', 'max', 'sum'])

print("Store Performance Stats:\n", store_stats)
```
