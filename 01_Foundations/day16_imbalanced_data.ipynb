#Handling Imbalanced Data (SMOTE & Sub-sampling)

Goal: Fix datasets where one class dominates the other (common in fraud detection or medical diagnosis).

#The Setup (Imports & Data)

This cell creates a "broken" dataset where the 0s outnumber the 1s 9-to-1.

```python
import pandas as pd
import numpy as np
from sklearn.utils import resample
from imblearn.over_sampling import SMOTE

# 1. Create fake imbalanced data
# 100 rows total: 90 zeros, 10 ones
X = np.random.rand(100, 2) 
y = np.array([0]*90 + [1]*10) 

# Put it in a DataFrame so we can see it clearly
df = pd.DataFrame(X, columns=['Feature_A', 'Feature_B'])
df['Target'] = y

print("Original Data Counts:")
print(df['Target'].value_counts())
```

#Manual Over-sampling (The "Copy-Paste" Method)

This is the simplest way to fix imbalance: you just take the 10 "ones" and duplicate them until you have 90 of them.

```python
# Separate the classes
df_majority = df[df['Target'] == 0]
df_minority = df[df['Target'] == 1]

# Upsample the minority
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # allows duplicating rows
                                 n_samples=90,    # match the majority count
                                 random_state=42)

# Combine back together
df_balanced = pd.concat([df_majority, df_minority_upsampled])

print("Balanced Data Counts (Manual):")
print(df_balanced['Target'].value_counts())
```

#SMOTE (The "Smart" Method)

SMOTE is better because it doesn't just copy old data; it looks at the minority points and creates new, synthetic points in the empty spaces between them.

```python
# Initialize SMOTE
sm = SMOTE(random_state=42)

# Create the new synthetic data
X_res, y_res = sm.fit_resample(X, y)

print(f"Old Shape: {X.shape}")
print(f"New Shape after SMOTE: {X_res.shape}")
print("New Class Counts:", np.bincount(y_res))
```
