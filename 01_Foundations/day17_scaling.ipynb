#Feature Scaling (Standardization vs. Normalization)

Goal: Ensure all features are on a similar scale so the model doesn't favor large numbers (like Salary) over small ones (like Age).

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np

# Sample Data: [Age, Income]
data = np.array([[25, 50000], [50, 200000], [35, 80000], [40, 120000]])

# 1. StandardScaler (Mean=0, Std=1) - Best for SVM, KNN, Linear Regression
std_scaler = StandardScaler()
standardized = std_scaler.fit_transform(data)

# 2. MinMaxScaler (Range 0 to 1) - Best for Neural Networks
minmax_scaler = MinMaxScaler()
normalized = minmax_scaler.fit_transform(data)

print("Standardized:\n", standardized)
print("\nNormalized:\n", normalized)
```
