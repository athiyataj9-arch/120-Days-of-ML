#Linear Algebra (Matrix Math)

Goal: Understand how Neural Networks "think" via matrix multiplication.

```python
import numpy as np

# Dot Product (Fundamental for Neural Network layers)
vec_a = np.array([1, 2, 3])
vec_b = np.array([4, 5, 6])
dot_product = np.dot(vec_a, vec_b)

# Matrix Multiplication (The weights logic)
matrix_a = np.random.rand(2, 3)
matrix_b = np.random.rand(3, 2)
result = np.matmul(matrix_a, matrix_b)

print("Dot Product:", dot_product)
print("Matrix Product Shape:", result.shape)
```
