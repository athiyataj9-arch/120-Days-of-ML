#Statistics for Machine Learning

Goal: Understand Central Tendency and Dispersion—the foundation of how models "evaluate" error.

#Descriptive Statistics

```python
import numpy as np
from scipy import stats

scores = np.array([85, 90, 78, 92, 88, 100, 50, 88]) # 50 is an outlier

print(f"Mean (Average): {np.mean(scores)}")
print(f"Median (Middle): {np.median(scores)}")
print(f"Mode (Frequent): {stats.mode(scores, keepdims=True).mode[0]}")
```

#Variance and Standard Deviation

This tells you how "spread out" your data is from the mean.

```python
# Standard Deviation: How much do scores vary?
std_dev = np.std(scores)
variance = np.var(scores)

print(f"Standard Deviation: {std_dev:.2f}")
print(f"Variance: {variance:.2f}")

# Z-Score: How many standard deviations is a point from the mean?
z_scores = stats.zscore(scores)
print(f"Z-Scores: {z_scores}")
```
