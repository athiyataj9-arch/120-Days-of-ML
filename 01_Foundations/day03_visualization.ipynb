#Exploratory Data Analysis (EDA)

Goal: Use Matplotlib and Seaborn to visualize distributions and correlations.

#Setup and Distribution Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Use the cleaned data from Day 2 or generate simple points
np.random.seed(42)
ages = np.random.normal(30, 5, 100)  # Mean 30, Std Dev 5

# Histogram: See the "shape" of data
plt.figure(figsize=(8, 4))
sns.histplot(ages, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()
```

#Relationships (Scatter & Heatmaps)

```python
# Create a correlation matrix
data = pd.DataFrame({
    'Experience': np.arange(10),
    'Salary': [40, 45, 50, 58, 65, 70, 78, 85, 90, 100]
})

# Heatmap: Find which features are strongly linked
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
