#Array Basics & Reshaping

Goal: Understand how data is structured for Machine Learning.

```python
import numpy as np

# 1. Create a 1D array of 12 elements
arr = np.arange(12)

# 2. Reshape into a 2D matrix (4 rows, 3 columns)
# This is how feature matrices are usually structured
matrix = arr.reshape(4, 3)

print("Original Array:", arr)
print("\n4x3 Matrix:\n", matrix)
print(f"\nShape: {matrix.shape} | Dimensions: {matrix.ndim}")
```

#Vectorization & Broadcasting

Goal: See how NumPy handles math efficiently without loops.

```python
# Create a 2D matrix
data = np.array([[1, 2, 3], [4, 5, 6]])

# Scalar Broadcasting: Add 10 to every single element
plus_ten = data + 10

# Row Broadcasting: Add [100, 200, 300] to each row
row_modifier = np.array([100, 200, 300])
broadcasted_sum = data + row_modifier

print("Original Data:\n", data)
print("\nAfter Scalar Addition (+10):\n", plus_ten)
print("\nAfter Row-wise Addition:\n", broadcasted_sum)
```

#Aggregations (Axis Logic)

Goal: Learn to summarize data by rows (samples) or columns (features).

```python
# Simulating 3 students' scores in 2 subjects
scores = np.array([
    [85, 90], 
    [70, 80], 
    [95, 85]
])

# Axis 0: Columns (Average per subject)
subject_avg = np.mean(scores, axis=0)

# Axis 1: Rows (Average per student)
student_avg = np.mean(scores, axis=1)

print("Scores Matrix:\n", scores)
print("\nAverage per Subject (Axis 0):", subject_avg)
print("Average per Student (Axis 1):", student_avg)
```
