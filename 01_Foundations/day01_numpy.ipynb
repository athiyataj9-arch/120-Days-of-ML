Advanced Array Creation & Shaping

#Goal: Moving beyond simple lists to generating structured data.

```python
import numpy as np

# 1. Generating data (Essential for ML initialization)
# Create a 1D array of 12 elements from 0 to 11
arr = np.arange(12)

# 2. Reshaping into 3D (Think of this as 2 images of 3x2 pixels)
# -1 allows NumPy to calculate the remaining dimension automatically
matrix_3d = arr.reshape(2, 3, 2) 

print(f"3D Shape: {matrix_3d.shape}")
print(f"Total Elements: {matrix_3d.size}")
print(f"Data Type: {matrix_3d.dtype}")
print(f"Array:\n{matrix_3d}")
```

Randomness & Initialization

#Goal: Learning how to create "Weights" for models.

```python
# Create a 3x3 matrix of random numbers between 0 and 1
random_matrix = np.random.rand(3, 3)

# Create a 3x3 identity matrix (Used in Linear Algebra)
identity = np.eye(3)

print("Random Weights:\n", random_matrix)
print("\nIdentity Matrix:\n", identity)
```

Vectorization & Broadcasting (The "Speed" Test)

#Goal: Understanding why we never use for loops in NumPy.

```python
# Create a large array to test speed
large_arr = np.arange(1_000_000)

# Vectorized Operation (Lightning fast)
%timeit large_arr * 2

# Broadcasting a 1D array across a 2D matrix
matrix = np.array([[1, 2, 3], [4, 5, 6]])
row_vector = np.array([10, 20, 30])

# Every row in the matrix gets the row_vector added to it
result = matrix + row_vector

print("Broadcasting Result:\n", result)
```

Axis-Based Aggregations

#Goal: Learning how to calculate statistics for specific columns or rows.

```python
data = np.array([
    [10, 20, 30], # Student 1 scores
    [40, 50, 60]  # Student 2 scores
])

# Axis 0 = Vertical (Columns) | Axis 1 = Horizontal (Rows)
column_means = np.mean(data, axis=0) # Mean score per subject
row_means = np.mean(data, axis=1)    # Mean score per student

print("Mean per Subject (Axis 0):", column_means)
print("Mean per Student (Axis 1):", row_means)
```
