Array Basics & Shapes

```python
import numpy as np
# Create a 1D array
arr_1d = np.array([1, 2, 3, 4, 5, 6])

# Create a 2D matrix (3 rows, 2 columns)
matrix = arr_1d.reshape(3, 2)

print(f"Shape: {matrix.shape}")
print(f"Dimensions: {matrix.ndim}")
print(f"Matrix:\n{matrix}")
```

Vectorization & Broadcasting (The Speed Factor)

```python
# Instead of a for loop, do math on the whole array at once
arr = np.array([10, 20, 30])

# Broadcasting: Adding a scalar (5) to every element
print("Addition:", arr + 5)

# Element-wise multiplication
print("Multiplication:", arr * 2)

# Aggregation functions
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
```
