#Hypothesis Testing (A/B Testing)

Goal: Mathematically prove if a change (like a new model feature) actually improved results or if the improvement was just random noise.

#The T-Test & P-Value

Practitioner Rule: If the P-Value < 0.05, the result is "Statistically Significant."

```python
from scipy import stats

# Conversion rates for 'Old Feature' vs 'New Feature'
old_feature = [0.12, 0.15, 0.14, 0.10, 0.13]
new_feature = [0.18, 0.20, 0.19, 0.17, 0.21]

# Perform Independent T-Test
t_stat, p_value = stats.ttest_ind(old_feature, new_feature)

print(f"P-Value: {p_value:.5f}")
if p_value < 0.05:
    print("Conclusion: Significant improvement! Deploy the new feature.")
else:
    print("Conclusion: Not significant. Stick with the old version.")
```
