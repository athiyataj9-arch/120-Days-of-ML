#SQL for Data (Pandas Integration)

Goal: Learn how to join and filter data, the industry standard for data extraction.

```python
import pandas as pd

# Creating two tables (Employees and Depts)
df_emp = pd.DataFrame({'ID': [1, 2], 'Dept_ID': [10, 20]})
df_dept = pd.DataFrame({'Dept_ID': [10, 20], 'Name': ['IT', 'HR']})

# SQL JOIN (Merge)
df_merged = pd.merge(df_emp, df_dept, on='Dept_ID', how='left')

print("Merged Data:\n", df_merged)
```
