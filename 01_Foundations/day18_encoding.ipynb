#Categorical Encoding

Goal: Convert text categories into numbers.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    'City': ['London', 'Paris', 'New York', 'Paris'],
    'Size': ['S', 'M', 'L', 'M']
})

# 1. One-Hot Encoding (Nominal - No order: City)
df_onehot = pd.get_dummies(df, columns=['City'])

# 2. Label Encoding (Ordinal - Natural order: Size)
le = LabelEncoder()
df['Size_Encoded'] = le.fit_transform(df['Size'])

print("One-Hot Encoded:\n", df_onehot)
print("\nLabel Encoded Size:", df['Size_Encoded'].values)
```
