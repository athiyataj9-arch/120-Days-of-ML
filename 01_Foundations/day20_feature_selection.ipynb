#Feature Selection (Correlation Analysis)

Goal: Identify and remove redundant features to simplify the model and prevent overfitting.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create data where Feature B is almost identical to Feature A
data = np.random.rand(100, 3)
df_corr = pd.DataFrame(data, columns=['Age', 'Salary', 'Redundant_Age'])
df_corr['Redundant_Age'] = df_corr['Age'] * 0.98 

# Plot Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title("Day 20: Correlation Heatmap")
plt.show()

# Insight: If correlation > 0.9, you can likely drop one feature.
```
