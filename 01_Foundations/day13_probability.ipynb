#Probability & Normal Distribution

Goal: Understand the "Bell Curve." Most Machine Learning models (like Linear Regression) assume your data or your errors follow this specific distribution.

#Visualizing the Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 random points (Mean=0, Std Dev=1)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot Histogram vs. Theoretical Bell Curve
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue')
x = np.linspace(-4, 4, 100)
plt.plot(x, norm.pdf(x, 0, 1), 'r-', lw=2)
plt.title("Normal Distribution (The Bell Curve)")
plt.show()
```

#Z-Scores (Finding Outliers)

A Z-score tells you how many standard deviations a point is from the mean. If $Z > 3$, it’s usually an outlier.

```python
def calculate_z_score(value, mean, std):
    return (value - mean) / std

# Example: If avg house price is 300k (std 50k), is a 500k house an outlier?
z = calculate_z_score(500, 300, 50)
print(f"Z-Score: {z}") # Output 4.0 (Extreme Outlier)
```
