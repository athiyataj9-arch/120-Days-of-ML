#Math for ML (Gradient Descent Concept)

Goal: Understand how models "learn" by minimizing error using derivatives.

```python
# Concept: Gradient Descent (Finding the minimum of a curve)
def f(x): return x**2  # Simple parabola

# Derivative (Slope)
def df(x): return 2 * x

x = 10 # Starting point (high error)
learning_rate = 0.1

# Take a step towards the minimum
x = x - learning_rate * df(x)
print(f"New position after one step: {x:.2f}")
```
