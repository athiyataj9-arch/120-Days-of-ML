#Matrix Inversion

Goal: Understand how to "undo" a matrix transformation. This is the foundation for solving linear equations in models like Linear Regression.

#The Determinant (The "Gatekeeper")

Note: Before inverting a matrix, you must check the determinant. If it is 0, the matrix is "singular" and cannot be inverted.

```python
import numpy as np

# A simple 2x2 Matrix
A = np.array([[4, 7], 
              [2, 6]])

# Calculate the Determinant
det = np.linalg.det(A)
print(f"Determinant: {det:.2f}")

if det == 0:
    print("Matrix is singular and cannot be inverted.")
```

#Calculating the Inverse

Concept: Multiplying a matrix by its inverse results in the Identity Matrix ($I$), which is the matrix equivalent of the number 1.

```python
# Calculate Inverse
A_inv = np.linalg.inv(A)
print("Inverse Matrix:\n", A_inv)

# Verification: A * A_inv should equal the Identity Matrix
identity = np.dot(A, A_inv)
print("\nVerification (Identity Matrix):\n", np.round(identity))
```
