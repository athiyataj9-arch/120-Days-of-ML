#K-Nearest Neighbors (KNN)

Goal: "Tell me who your neighbors are, and I'll tell you who you are." We classify a point by looking at the $K$ closest points around it.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# 1. Load Data
iris = load_iris()
X = iris.data[:, :2] # Only use 2 features so we can plot it
y = iris.target

# 2. KNN REQUIRES Scaling (Crucial Step!)
# Because KNN uses distance, we must make all units equal
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Train KNN with 5 neighbors
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_scaled, y)

# 4. Visualize the "Neighborhoods"
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title("Day 28: K-Nearest Neighbors (Clusters of Iris Flowers)")
plt.xlabel("Sepal Length (Scaled)")
plt.ylabel("Sepal Width (Scaled)")
plt.savefig('../assets/day28_plot.png')
plt.show()

print(f"Model accuracy on training data: {knn.score(X_scaled, y)*100:.2f}%")
```
