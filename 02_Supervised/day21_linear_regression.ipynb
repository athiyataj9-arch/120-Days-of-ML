#Simple Linear Regression (from scratch)

Goal: Understand the math behind the "Best Fit Line." We use the formula $y = mx + c$, where $m$ is the slope (weight) and $c$ is the intercept (bias).

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# 2. Math: Normal Equation (theta = (X^T . X)^-1 . X^T . y)
X_b = np.c_[np.ones((100, 1)), X] 
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# 3. Plot & Save (This replaces the stock images)
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, X_b.dot(theta_best), color='red', linewidth=2, label='Prediction')
plt.title("Day 21: Linear Regression Model")
plt.legend()
plt.savefig('../assets/day21_plot.png') # Saves the file for your README
plt.show()
```
