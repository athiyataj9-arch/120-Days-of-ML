#Gradient Boosting Machine (GBM)

**Logic:** Instead of just changing weights, GBM tries to predict the residuals (the errors) of the previous model using Gradient Descent. It "nudges" the prediction closer to the truth.

#Setup Data

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

#Training

```python
gbm = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbm.fit(X_train, y_train)

print(f"GBM Test Accuracy: {gbm.score(X_test, y_test)*100:.2f}%")
```
