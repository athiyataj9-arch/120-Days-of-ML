#Decision Trees

**Logic:** "The Flowchart Model." It asks a series of "If/Else" questions to split the data into smaller, purer groups.

#Imports & Model Training

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
# max_depth=3 prevents the tree from becoming too complex (overfitting)
dt_clf = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_clf.fit(iris.data, iris.target)
```

#Visualizing the Tree Logic

```python
plt.figure(figsize=(15, 10))
plot_tree(dt_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Day 31: Decision Tree Flowchart")
plt.savefig('../assets/day31_plot.png')
plt.show()
```
