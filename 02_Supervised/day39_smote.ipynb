#Handling Imbalanced Data (SMOTE)

**The Goal:** In the real world, "Fraud" or "Disease" cases are rare (1%). If the model just guesses "No Disease" 100% of the time, it's 99% accurate but useless. SMOTE creates synthetic "Fraud" cases to balance the scales.

Note: Run pip install imbalanced-learn in your terminal.

#Logic and Synthetic Sampling

```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
from collections import Counter

# Create highly imbalanced data (90% vs 10%)
X, y = make_classification(n_classes=2, weights=[0.9, 0.1], n_samples=1000, random_state=42)
print(f"Original distribution: {Counter(y)}")

# Synthesize new examples for the minority class
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

print(f"Balanced distribution: {Counter(y_res)}")
```
