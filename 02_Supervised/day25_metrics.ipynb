#Regression Evaluation Metrics

Goal: In plain English, we are asking: "On average, how far off are our predictions from the truth?"

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 1. Create Sample Data (Actual vs Predicted)
# Imagine these are actual house prices vs what our model guessed
y_true = np.array([300, 450, 500, 610, 580]) 
y_pred = np.array([310, 440, 490, 650, 570]) 

# 2. Calculate the "Big Three" Metrics
mae = mean_absolute_error(y_true, y_pred)   # Average error (simple)
mse = mean_squared_error(y_true, y_pred)    # Error squared (punishes big misses)
rmse = np.sqrt(mse)                         # Error in original units ($)
r2 = r2_score(y_true, y_pred)               # Accuracy score (1.0 is perfect)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R-squared (Model Fit): {r2:.2f}")

# 3. Visualization: The Residual Plot
# We want to see if our errors are random or if there's a pattern we missed.
residuals = y_true - y_pred
plt.figure(figsize=(8, 4))
plt.axhline(0, color='red', linestyle='--')
plt.scatter(y_pred, residuals, color='purple')
plt.title("Residual Plot: Are our errors random?")
plt.xlabel("Predicted Values")
plt.ylabel("Errors (Actual - Predicted)")
plt.savefig('../assets/day25_plot.png')
plt.show()
```
