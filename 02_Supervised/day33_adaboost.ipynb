#AdaBoost (Adaptive Boosting)

**Logic:** It’s like a teacher who gives extra homework only on the questions you got wrong. It focuses more weight on misclassified points in every round.

#Imports & Training

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

#Model & Plot

```python
ada = AdaBoostClassifier(n_estimators=50, random_state=42)
ada.fit(X_train, y_train)

print(f"AdaBoost Accuracy: {ada.score(X_test, y_test)*100:.2f}%")
```
