#Regularization (Ridge & Lasso)

Goal: Prevent overfitting by adding a penalty to the size of the coefficients.

- Ridge ($L_2$): Keeps weights small.
- Lasso ($L_1$): Can force some weights to exactly 0 (helps in feature selection).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso

# 1. Generate noisy linear data (Defining X and y)
np.random.seed(42)
X = 3 * np.random.rand(100, 1)
y = 1 + 0.5 * X + np.random.randn(100, 1) / 1.5

# 2. Ridge Regression (L2 Regularization)
ridge_reg = Ridge(alpha=1, solver="cholesky")
ridge_reg.fit(X, y)

# 3. Lasso Regression (L1 Regularization)
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)

# 4. Predictions for plotting
X_new = np.linspace(0, 3, 100).reshape(100, 1)
y_ridge = ridge_reg.predict(X_new)
y_lasso = lasso_reg.predict(X_new)

# --- 5. THE "TEXTBOOK STYLE" PLOT ---
fig, ax = plt.subplots(figsize=(8, 5))

# Move axes to center
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Plot Data and Models
plt.scatter(X, y, color='red', s=20, alpha=0.5, label='Data Points')
plt.plot(X_new, y_ridge, color='blue', linewidth=2, label='Ridge (L2)')
plt.plot(X_new, y_lasso, color='green', linewidth=2, linestyle='--', label='Lasso (L1)')

plt.title("Day 23: Ridge vs Lasso Regularization", pad=20)
plt.legend()

# Save to assets
plt.savefig('../assets/day23_plot.png', bbox_inches='tight')
plt.show()

print(f"Ridge Coefficients: {ridge_reg.coef_}")
print(f"Lasso Coefficients: {lasso_reg.coef_}")
```
