#Activation Functions (Relu, Sigmoid, Softmax)

**Logic:** Activation functions are like "filters."

- **Sigmoid:** Squashes numbers between 0 and 1 (Probability).

- **ReLU:** Turns negative numbers into 0 (The most popular choice).

- **Softmax:** Used at the very end to pick 1 winner out of many categories.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x): return np.maximum(0, x)
def sigmoid(x): return 1 / (1 + np.exp(-x))

x = np.linspace(-5, 5, 100)

plt.plot(x, relu(x), label='ReLU')
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.title("Day 58: Visualizing Activation Functions")
plt.legend()
plt.grid(True)
plt.savefig('../../assets/day58_activations.png')
plt.show()
```
