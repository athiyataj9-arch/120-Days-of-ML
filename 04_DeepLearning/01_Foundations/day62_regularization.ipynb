#Regularization (Dropout & L2)

**Logic:** We add Dropout. It’s like a teacher telling students they can't use their notes for one specific chapter—it forces them to actually understand the concepts rather than just memorizing the page numbers.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2

# 1. Build a "Disciplined" Model
# We use 10 input features as an example
model = Sequential([
    # Layer 1: 64 neurons with L2 regularization (penalty for huge weights)
    Dense(64, activation='relu', kernel_regularizer=l2(0.01), input_dim=10),
    
    # Dropout: 30% of neurons sleep during each training step
    Dropout(0.3), 
    
    # Layer 2: 32 neurons
    Dense(32, activation='relu'),
    
    # Dropout: 20% of neurons sleep here
    Dropout(0.2), 
    
    # Output Layer: Binary Classification
    Dense(1, activation='sigmoid')
])

# 2. Compile
model.compile(
    optimizer='adam', 
    loss='binary_crossentropy', 
    metrics=['accuracy']
)

# 3. Check the Architecture
print("Day 62: Regularization Model Summary")
model.summary()
```
