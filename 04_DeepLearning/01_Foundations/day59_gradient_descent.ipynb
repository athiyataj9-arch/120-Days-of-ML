#Gradient Descent & Backpropagation

**Logic:** If the network makes a mistake, how does it know which weight to change? Gradient Descent is like walking down a mountain in the fog—you feel the slope under your feet and move toward the lowest point (the lowest error).

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. A simple "Loss" function: y = x^2
def loss_function(x): return x**2
def gradient(x): return 2*x  # The "Slope"

# 2. Start at a random "Weight"
weight = 4
learning_rate = 0.1

# 3. Take 10 steps down the mountain
history = []
for _ in range(10):
    history.append(weight)
    weight = weight - (learning_rate * gradient(weight))

print(f"Final Optimized Weight: {weight:.4f} (Close to 0)")

# 4. Save visualization
plt.plot(np.linspace(-5, 5), loss_function(np.linspace(-5, 5)))
plt.scatter(history, [loss_function(h) for h in history], color='red')
plt.title("Day 59: Stepping Down the Loss Curve")
plt.savefig('../../assets/day59_plot.png')
plt.show()
```
