#Overfitting vs. Underfitting

**Logic:** 

- **Overfitting:** The model is so smart it memorizes the training data but fails in the real world (High variance).

- **Underfitting:** The model is too simple and learns nothing (High bias).

```python
# Usually visualized via training curves. 
# If 'Val_Loss' starts going UP while 'Train_Loss' goes DOWN, you are overfitting!
print("Day 61 concept: Validation curves are the pulse of your model.")
```
