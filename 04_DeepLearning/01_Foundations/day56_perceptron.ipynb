#Introduction to Neural Networks (The Perceptron)

**Logic:** A Neural Network is just a math version of a brain cell. It takes inputs, gives them "weights" (importance), adds them up, and decides if it should "fire" (output 1 or 0).

```python
import numpy as np

# 1. Inputs (e.g., Is it sunny? Is it a weekend?)
inputs = np.array([1, 0]) 

# 2. Weights (How much we care about each input)
weights = np.array([0.7, 0.3])

# 3. Bias (Our internal threshold/preference)
bias = -0.5

# 4. Calculation: (Input * Weight) + Bias
step1 = np.dot(inputs, weights) + bias

# 5. Activation: If > 0, output 1 (Go for a walk), else 0 (Stay home)
output = 1 if step1 > 0 else 0

print(f"Perceptron Decision: {output}")
```
