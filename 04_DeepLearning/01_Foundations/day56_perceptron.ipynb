#Introduction to Neural Networks (The Perceptron)

**Logic:** The Perceptron is a linear classifier. It tries to draw a straight line to separate "Yes" from "No." We will plot the inputs and show where the neuron makes its decision.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Inputs and Weights
inputs = np.array([1, 0.5]) 
weights = np.array([0.7, -0.5])
bias = -0.2

# 2. The Perceptron Math
summation = np.dot(inputs, weights) + bias
output = 1 if summation > 0 else 0

# 3. Visualization of the "Input Space"
plt.figure(figsize=(6, 4))
plt.scatter(inputs[0], inputs[1], color='red', s=100, label=f'Input Point (Result: {output})')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("Day 56: Perceptron Input Visualization")
plt.grid(True, linestyle='--')
plt.legend()

# 4. Save to assets
plt.savefig('../../assets/day56_plot.png')
plt.show()
```
