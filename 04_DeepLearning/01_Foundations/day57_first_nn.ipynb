#Your First Neural Network (Keras Sequential)

**Logic:** We use a library to build a "Stack" of layers. We’ll try to solve a simple classification problem (like the XOR problem or a small dataset).

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# 1. Simple Data: Binary Inputs (X) and Target (y)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 1, 1, 0]) # XOR Logic

# 2. Build the "Brain"
model = Sequential([
    Dense(4, input_dim=2, activation='relu'), # Hidden layer with 4 neurons
    Dense(1, activation='sigmoid')            # Output layer (Binary)
])

# 3. Compile (Define the 'Loss' and 'Optimizer')
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. Train
model.fit(X, y, epochs=100, verbose=0)
print("Model trained on XOR logic!")
```
