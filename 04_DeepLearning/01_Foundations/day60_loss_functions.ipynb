#Loss Functions (MSE vs. Cross-Entropy)

**Logic:** The "Loss" is the penalty for being wrong.

- **MSE (Mean Squared Error):** Best for predicting numbers (Price, Age).

- **Binary Cross-Entropy:** Best for Yes/No questions (Cat vs. Dog).

```python
import tensorflow as tf

# 1. Regression Loss (Guessing a price)
y_true_reg = [100.0]
y_pred_reg = [90.0]
mse = tf.keras.losses.MeanSquaredError()
print(f"Regression Error (MSE): {mse(y_true_reg, y_pred_reg).numpy()}")

# 2. Classification Loss (Is it a Dog?)
y_true_cls = [[1.0]] 
y_pred_cls = [[0.8]] 

# --- THE FIX: lowercase 'e' in Crossentropy ---
bce = tf.keras.losses.BinaryCrossentropy() 
print(f"Classification Error (BCE): {bce(y_true_cls, y_pred_cls).numpy()}")
```
