#Content-Based Filtering

**Logic:** "If you liked this, you'll like that." We compare item features (tags). We use Cosine Similarity to see how similar two items are. 1.0 means identical, 0.0 means totally different.

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# 1. Movie Profiles: [Action, Sci-Fi, Comedy]
movies = {
    'Terminator': [1, 1, 0], 
    'Star Wars':  [1, 1, 0], 
    'Toy Story':  [0, 0, 1]
}
df = pd.DataFrame(movies).T

# 2. Calculate how similar they are to each other
sim_matrix = cosine_similarity(df)
sim_df = pd.DataFrame(sim_matrix, index=df.index, columns=df.index)

print("How similar are these movies?")
print(sim_df)
```
