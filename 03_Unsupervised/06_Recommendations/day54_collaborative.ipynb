#Collaborative Filtering

**Logic:** "Users like you also enjoyed..." We look at the correlation between users. If User A and User B both gave 5 stars to the same 3 movies, they have similar taste.

```python
import pandas as pd

# Ratings given by Users for different genres (1 to 5 stars)
data = {
    'User_A': [5, 4, 1], # Likes Action/Sci-Fi
    'User_B': [5, 5, 2], # Likes Action/Sci-Fi
    'User_C': [1, 2, 5]  # Likes Romance
}
df_ratings = pd.DataFrame(data, index=['Action', 'SciFi', 'Romance'])

# Find the correlation between users' tastes
user_correlation = df_ratings.corr()

print("User Taste Similarity:")
print(user_correlation)
```
