#Recommendation System Capstone

**Logic:** We aren't just looking at numbers anymore; we are visualizing "Taste Clusters." The heatmap shows exactly which users are "twins" in their movie preferences.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create a Rating Matrix (Users x Movies)
ratings_dict = {
    'User_1': [5, 4, 1, 1],
    'User_2': [4, 5, 1, 2],
    'User_3': [1, 1, 5, 4],
    'User_4': [2, 1, 4, 5]
}
movies = ['Action_A', 'Action_B', 'Romance_A', 'Romance_B']
df = pd.DataFrame(ratings_dict, index=movies).T

# 2. Calculate User Similarity (Correlation)
user_sim = df.T.corr()

# 3. Visualize the "Taste Match"
plt.figure(figsize=(8, 6))
sns.heatmap(user_sim, annot=True, cmap='YlGnBu', cbar=True)
plt.title("Day 55: User Similarity (Recommendation Logic)")

# 4. Save to assets
plt.savefig('../../assets/day55_plot.png')
plt.show()
```
