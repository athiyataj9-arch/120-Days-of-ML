#Recommendation System Capstone

**Logic:** We combine everything. We take a dataset of users and movies, calculate a "Similarity Score," and then predict what a user would like based on what similar users enjoyed.

```python
import pandas as pd

# 1. Create a Rating Matrix (Users x Movies)
# NaN means the user hasn't watched that movie yet
ratings_dict = {
    'User_1': [5, 4, 1, 1],
    'User_2': [4, 5, 1, 2],
    'User_3': [1, 1, 5, 4],
    'User_4': [2, 1, 4, 5]
}
movies = ['Action_A', 'Action_B', 'Romance_A', 'Romance_B']
df = pd.DataFrame(ratings_dict, index=movies).T

# 2. Find Correlation (Who has similar taste?)
user_sim = df.T.corr()

# 3. Recommend: If User_1 likes Action_A, find who else likes it and what they watched.
print("User Similarity (Taste Match):")
print(user_sim)

# Logic: User 1 and User 2 are highly correlated (Action fans)
# User 3 and User 4 are highly correlated (Romance fans)
```
