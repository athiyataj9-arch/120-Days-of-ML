#K-Means Clustering (The Basics)

**Logic:** Imagine you have a crowd of people and you want to split them into 3 groups based on where they are standing. K-Means picks 3 random spots (centroids), assigns people to the closest spot, and then moves the spots to the center of those groups until they stop moving.

1. Imports & Synthetic Data

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Create 300 data points with 4 centers (clusters)
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualize the raw data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Day 41: Raw Unlabeled Data")
plt.show()
```

2. Applying K-Means

```python
# Initialize K-Means with 4 clusters
kmeans = KMeans(n_clusters=4, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Plot the clusters and their centroids
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("Day 41: K-Means Clustering Result")
plt.savefig('../../assets/day41_plot.png')
plt.show()
```
