#DBSCAN (Density-Based Clustering)

**Logic:** K-Means fails if the clusters are weird shapes (like circles inside circles). DBSCAN finds "high-density" areas. It’s like finding a city: it doesn't care about the shape, just that the people are packed closely together. It also automatically finds outliers (noise).

**Goal:** Finding clusters by how "packed" the points are. Great for weird shapes.

1. Setup "Moon" Data

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Create moon-shaped data (K-Means fails on this!)
X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)
```

2. Run DBSCAN

```python
# eps = radius to look for neighbors; min_samples = points needed to be a "cluster"
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma')
plt.title("Day 44: DBSCAN (Density Clustering)")
plt.savefig('../../assets/day44_plot.png')
plt.show()
```
