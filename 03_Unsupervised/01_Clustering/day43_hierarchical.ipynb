#Hierarchical Clustering (Dendrograms)

**Logic:** Think of this like a family tree. It starts with every data point as its own cluster and slowly merges the "closest relatives" together until everyone is in one big group.

**Goal**: Building a "Family Tree" (Dendrogram) for your data.

1. Setup Data

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

# Create the data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

2. Draw Dendrogram & Cluster

```python
# 1. Draw the Dendrogram (The "Tree")
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Day 43: Dendrogram')
plt.savefig('../../assets/day43_dendrogram.png')
plt.show()

# 2. Apply the actual clustering
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_hc, cmap='rainbow')
plt.title('Day 43: Hierarchical Clusters')
plt.show()
```
