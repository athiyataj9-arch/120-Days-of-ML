#ECLAT Algorithm (Simple & Fast)

**Logic:** If Apriori is about "Rules," ECLAT is about "Popularity." It doesn't care about Confidence or Lift. It simply looks at itemsets and says: "These things appear together X% of the time." It’s much faster for smaller datasets because it uses a vertical data format.

Implementation

```python
import pandas as pd

# 1. We recreate the data here so it's defined in this notebook
# 1 = Item was in the cart, 0 = It was not
data = {
    'Milk':   [1, 0, 1, 1, 0], 
    'Bread':  [1, 1, 1, 1, 1], 
    'Butter': [1, 1, 0, 1, 1]
}
df = pd.DataFrame(data)

# 2. ECLAT Logic: Calculate Support (Frequency)
def run_eclat(dataframe, min_support):
    # Support is just the average of the 1s and 0s
    supports = dataframe.mean().sort_values(ascending=False)
    # Filter by the threshold
    return supports[supports >= min_support]

print("ECLAT Support Scores (Popularity):")
print(run_eclat(df, 0.5))
```
