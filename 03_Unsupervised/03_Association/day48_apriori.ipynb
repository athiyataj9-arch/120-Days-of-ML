#Market Basket Analysis (Apriori)

**Logic:** This is the "Bread & Butter" algorithm. It looks at thousands of receipts to find a rule: If a customer buys A, how likely are they to buy B? We use Support (how common the combo is) and Confidence (how certain the rule is).

Imports

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

Easy Rule Mining

```python
# 1. Simple shopping list
data = [['Milk', 'Bread'], ['Bread', 'Butter'], ['Milk', 'Butter'], ['Milk', 'Bread', 'Butter']]

# 2. Convert to table (One-Hot Encoding)
te = TransactionEncoder()
encoded_data = te.fit(data).transform(data)
df = pd.DataFrame(encoded_data, columns=te.columns_)

# 3. Find popular combinations (Support)
freq_items = apriori(df, min_support=0.5, use_colnames=True)

# 4. Find the rules (Confidence)
rules = association_rules(freq_items, metric="confidence", min_threshold=0.7)

print("Shopping Patterns Found:")
print(rules[['antecedents', 'consequents', 'support', 'confidence']])
```
