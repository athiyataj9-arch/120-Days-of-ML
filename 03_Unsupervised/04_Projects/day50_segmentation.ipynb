#Customer Segmentation Capstone

**Logic:** We take a messy dataset with 10 features. We use PCA to simplify it into 2 "main behaviors" so we can see it, then use K-Means to label customers as "Groups."

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Create data: 500 customers with 10 different buying habits (features)
X, _ = make_blobs(n_samples=500, n_features=10, centers=4, random_state=42)

# 2. PCA: "Squash" the 10 features into 2 Principal Components so we can see them
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 3. K-Means: Find the 4 groups within that simplified data
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_pca)

# 4. Plot the result
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title("Day 50: Customer Segments (PCA + K-Means)")
plt.xlabel("Behavior Component 1")
plt.ylabel("Behavior Component 2")
plt.show()
```
