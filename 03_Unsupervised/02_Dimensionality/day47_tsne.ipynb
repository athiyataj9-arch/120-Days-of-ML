#t-SNE (Non-Linear Compression)

**Logic:** PCA is linear. t-SNE is a "flexible" compression tool. It’s amazing for visualizing complex data clusters (like hand-written digits) in 2D.

1. Self-Contained Implementation

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

```python
# 1. Load Data
digits = load_digits()
```

```python
# 2. Run t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(digits.data)
```

```python
# 3. Plot
plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=digits.target, cmap='tab10', s=10)
plt.colorbar()
plt.title("Day 47: t-SNE visualizing 64-dimensional Digits")
```
