#PCA (Mathematics & Variance)

**Logic:** PCA finds the "directions" (Principal Components) where the data varies the most. Instead of 10 features, we find the 2 directions that explain 95% of the information.

1. Setup & Manual Variance Check

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load Iris - 4 features (Hard to visualize in 4D!)
data = load_iris()
X = data.data
```

2. Dimensionality Reduction

```python
# Reduce from 4D to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data.target)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Day 45: 4D data reduced to 2D via PCA')
plt.savefig('../../assets/day45_plot.png')
plt.show()
```
