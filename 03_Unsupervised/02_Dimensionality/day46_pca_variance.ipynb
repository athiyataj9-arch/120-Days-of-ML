#PCA Explained Variance Ratio

**Logic:** How do we know if 2 components are enough? We check the "Explained Variance Ratio." It tells us exactly how much "truth" we kept during the compression.

1. Setup Data & PCA

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# 1. Load 4D data
data = load_iris()
X = data.data
```

```python
# 2. Reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# 3. Plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data.target)
plt.title('Day 45: PCA (4D reduced to 2D)')
plt.savefig('../../assets/day45_plot.png')
plt.show()
```
