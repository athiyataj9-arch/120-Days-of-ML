#Local Outlier Factor (LOF)

**Logic:** This looks at Density. If everyone else is standing in a tight circle and you are standing 10 feet away, your "Local Outlier Factor" is high.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# 1. Define Data (Self-contained)
X_normal = 0.3 * np.random.randn(200, 2)
X_outliers = np.random.uniform(low=-4, high=4, size=(10, 2))
X = np.vstack([X_normal, X_outliers])

# 2. Local Outlier Factor Logic
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# 3. Visualization
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolors='k')
plt.title("Day 52: Local Outlier Factor (Density Detection)")

# --- THE FIX: Save BEFORE showing ---
plt.savefig('../../assets/day52_plot.png') 
plt.show()
```
