#Isolation Forest (Anomaly Detection)

**Logic:** Anomalies are "lonely" points. Isolation Forest works by randomly splitting the data. Because outliers are far away from the crowd, they get isolated very quickly (fewer splits).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 1. Create data: 200 normal points + 10 "fraud" points (outliers)
X_normal = 0.3 * np.random.randn(200, 2)
X_outliers = np.random.uniform(low=-4, high=4, size=(10, 2))
X = np.vstack([X_normal, X_outliers])

# 2. Identify the anomalies (expecting 5% of data to be weird)
iso = IsolationForest(contamination=0.05, random_state=42)
preds = iso.fit_predict(X) # 1 = Normal, -1 = Anomaly

# 3. Plot (Red points are the anomalies)
plt.scatter(X[:, 0], X[:, 1], c=preds, cmap='RdYlGn')
plt.title("Day 51: Finding Anomalies with Isolation Forest")
plt.show()
```
